### boxplot of VehOdo



```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=auction['VehOdo'])
plt.title('Boxplot of Vehicle Odometer Readings')
plt.show()
```

```python
auction.VehOdo.plot(kind='box',title='Boxplot of mileage in the car data set')
plt.plot()
plt.show()
```

#### These box plots are showing the distribution of the mileage for the sample of cars, that we have data on. We can see where the data is distributed, the avaerage, where most of the data is located in between, or the quartiles, we can see the minimum and maximum values for the data, and we can also see the outliers or data that is not within the normal bounds for the data population.As shown, we can see the average car mileage is 70,000 with almost 90% of the cars within 110,000 and 30,000 miles. it seems to be very unusual or uncommon for a car in this dataset to be under 30,000 according to this dataset

## Barplot of count by size

```python
auction['Size'].value_counts().plot(kind='bar')
plt.title('Count of Vehicles by Size')
plt.xlabel('Size')
plt.ylabel('Count')
plt.show()
```

#### this bar plot is ahowing us the popularities of vehicles in aech size category, showing us that medium cars are by far the most popular car and hold signmificant share of the market. with the majority of cars being dpread oreety evenly among all types of cars, with a concentration on meduim.

```python
plt.figure(figsize=(10, 6))
plt.hist(auction['WarrantyCost'], bins=30)
plt.title('Distribution of Warranty Cost')
plt.xlabel('Warranty Cost')
plt.ylabel('Frequency')
plt.show()
```

#### this histogram is showing us a count each possible warranty cost and aggregating the results. this is showing us that on average, mist warranty costs tend to start around 500-1,500 with none going lower, and some posssiblilities to be higher than 1,500. The data is skewed higher meaning that there are more data points closer the mean that are smaller, and as warrannty costs go up, there might be a slim possibility of having a variable.

```python
sns.pairplot(auction[['Auction', 'Color', 'IsBadBuy', 'MMRCurrentAuctionAveragePrice', 'Size',
       'TopThreeAmericanName', 'VehBCost', 'VehicleAge', 'VehOdo',
       'WarrantyCost', 'WheelType']])
plt.suptitle('Pairplot of Car Auction Data', y=1.02)
plt.show()
```

```python
sns.pairplot(auction[['Auction', 'Color', 'IsBadBuy', 'MMRCurrentAuctionAveragePrice', 'Size',
       'TopThreeAmericanName', 'VehBCost', 'VehicleAge', 'VehOdo',
       'WarrantyCost', 'WheelType']], hue='IsBadBuy')
plt.suptitle('Pairplot of Car Auction Data (Hue: IsBadBuy)', y=1.02)
plt.show()
```
